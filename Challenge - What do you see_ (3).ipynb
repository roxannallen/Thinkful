```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('/Users/roxyallen/Downloads/predictblooddonations.csv')
```

```python
df.describe()
```

The below graph is a historgram of The Number of Donations that people have made to the blood bank. It is negatively skewed and has a long tail. You can easily see from this graph that most people make under 5 total donations.

```python
plt.hist(df['Number of Donations'], bins=60)

plt.xlabel('Number of Donations')
plt.ylabel('Frequency')
plt.title('Histogram of Number of Donations')
```

In this line graph, we have the number of donations versus the total volume donated in c.c.'s. You can easily see in this line graph that there is a positive, relatively strong, relationship between the two variables.

```python
x = df['Number of Donations']
y = df['Total Volume Donated (c.c.)']

plt.plot(x, y, color='purple')
plt.xlabel('Number of Donations')
plt.ylabel('Total Volume Donated (c.c.)')
plt.title('Number of Donations vs. Total Volume Donated in C.C.s')
plt.show()
```

Here we have two subplots that compare the average number of months since the first versus the last donation. You can see that the average number of months since the first donation is much higher than the average number of months since the last and that the average number of months since the last donation has some interesting outliers at the top.

```python
plt.figure(figsize=(10, 5))
x = df['Months since First Donation']
plt.subplot(1, 2, 1)
plt.boxplot(x)
plt.ylabel('Number of Months since First Donation')
plt.title('Average Number of Months since First Donation')

y = df['Months since Last Donation']
plt.subplot(1, 2, 2)
plt.boxplot(y)
plt.ylabel('Number of Months since Last Donation')
plt.title('Average Number of Months since Last Donation')
plt.show()
```

The below is a scatterplot of the months since the first donation versus the total number of donations. You can see that there is a positive relationship between the two variables, but the data is very spread out so there is likely a high standard deviation.

```python
plt.scatter(x=df['Months since First Donation'], y=df['Number of Donations'])
plt.xlabel('Months since First Donation')
plt.ylabel('Total Number of Donations')
plt.title('Months since First Donation vs. the Total Number of Donations')
plt.show()
```
