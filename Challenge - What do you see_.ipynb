```python
# The top left plot shows the frequency of the dollar amount of the goals Kickstarter users set for their projects for December 2018. 
# It is a nice visualization that immediately shows the peak at $1,000 for a project goal; and then peaks again at 
# just above $1,500 and just above $2,000. The most interesting is the peak at $1,000 in this visualization.

# The top right plot shows the frequency of the dollar amount of the goals Kickstarter users set for their projects for December 2018. 
# versus how much was pledged. You can quickly see the stark difference between the goal and pledges and that most don't reach their goal.

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['goal'])
plt.xlabel('Amount in Dollars')
plt.ylabel('Number of Kickstarter Organizers')
plt.title('Kickstarter Goals for December 2018')

plt.subplot(1, 2, 2)
plt.plot(df['goal'], color='purple')
plt.plot(df['usd_pledged'], color='green')
plt.xlabel('Amount in Dollars')
plt.ylabel('Number of Kickstarter Organizers')
plt.title('Goals in purple vs. Pledges in green')

plt.tight_layout()
plt.show()
```

```python
##In this scatterplot, you can see that there are not a large number of backers and that as the project goal
##increases, the number of backers does not increase. Maybe you shouldn't go big or go home? 
##You can also see a couple of outliers.

plt.scatter(x=df['backers_count'], y=df['goal'])
plt.xlabel('Number of Backers')
plt.ylabel('Project Goal')
plt.title('Kickstarter Backers vs. Goals for December 2018')
plt.show()
```

```python
# Compare the goal amount vs. backers count between the U.S. and Canada. It looks like Canada
# doesn't have the outliers that the U.S. does, there isn't quite as much spread in the data
# so goals and number of backers are matched better in Canada.

us = df[df.country == 'US']
canada = df[df.country == 'CA']
```

```python
plt.scatter(us.backers_count, us.goal)
plt.scatter(canada.backers_count, canada.goal)
plt.show()
```
