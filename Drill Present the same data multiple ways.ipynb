```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/roxyallen/Downloads/Kickstarter_2018-12-13T03_20_05_701Z/Kickstarter.csv')
```

```python
df.columns
```

```python
plt.subplots(figsize=(10,8))
sns.boxplot(df.usd_pledged)

plt.xlabel('Pledged Amount')
plt.title('Boxplot for Pledged Amount for Nov-Dec 2018')
plt.show()
#This is a boxplot of the pledged amount for this month. Pro: Can easily see outliers. Con: Difficult to understand the median since there are so many 0's in the data.
```

```python
plt.subplots(figsize=(10,8))
sns.distplot(df.usd_pledged[df.usd_pledged < 100], hist=False, rug=True, label='Pledged Amount').legend()

plt.xlabel('Pledged Amount')
plt.ylabel('Kernel Density Estimation')
plt.title('Pledged Amount Kernel Density Estimation')
plt.show()
#This is a Kernel Density Estimation plot. Pro: You can easily see where many of the data points are concentrated. Con: Not easy for the general public to interpret and make meaning from it.
```

```python
plt.subplots(figsize=(10,8))
sns.distplot(df.usd_pledged[df.usd_pledged < 100], label='Pledged Amount').legend()

plt.xlabel('Pledged Amount')
plt.ylabel('Normalized Distribution')
plt.title('Pledged Amount Distribution')
plt.show()
#This is a regular normal distrubtion of just the pledged amount in US Dollars under 100. Pro: easy to see where the data spikes. Con: With a lot of 0's hard to see the full distribution.
```

```python
plt.figure(figsize = (12, 8))
plt.scatter(range(df.shape[0]), np.sort(df.usd_pledged.values))
plt.xlabel('index', fontsize=12)
plt.ylabel('Pledged Amount', fontsize=12)
plt.title("Pledged Amount Distribution")
plt.show()
#This is a scatterplot of the pledged amount in US Dollars. Pro: You can easily see that much of the data are small amounts of pledges, with 1 major outlier. Con: May not be easy for some people to interpret just upon first glance.
```

```python
x=df.goal[df.goal < 100]
y=df.usd_pledged[df.usd_pledged < 100]
sns.scatterplot(x,y)
#This is a scatterplot of the relationship between the goal and the pledged amount in US Dollars. You can see there is no real relationship. Pro: Easily see relationships between 2 variables. Con: No line to better see any relationship.
```

```python
x=df.goal[df.goal < 100]
y=df.usd_pledged[df.usd_pledged < 100]
sns.lineplot(x,y)
#This is a line graph showing the relationship between the goal and pledged amount in US Dollars for those under 100. Pro: Can see any relationship with the line. Con: Can be hard to interpret if the line is not straight, like this one.
```

```python
sns.regplot(x="goal", y="usd_pledged", data=df)
#This is a regression plot to understand the relationship between the goal and pledged variables. Pro: The line is fitted and you can easily see if there's a relationship visually. Con: Doesn't tell you much else except the degree of the relationship.
```

```python
sns.catplot(x="staff_pick", y="usd_pledged", data=df)
#This categorical plot examines the pledged amount by whether or not the project was a staff pick for the month. Pro: You can easily see the differences in the concentration of variables and the outliers. Con: Could be more detail, especially where the concentration of variables is at the bottom.
```

```python
sns.catplot(x="staff_pick", y="usd_pledged", jitter=False, data=df);
#This plot takes out the "jitter" of the previous plot. Pro: The lines are straighter, and this would be nice for a more streamlined look, especially if you were comparing more variables. Con: Not as visually impactful as with the jitter.
```

```python
sns.catplot(x="staff_pick", y="usd_pledged", kind="box", data=df);
#This is a classic boxplot of the categorical variable. Pro: It presents the interquartile ranges and the median, unlike the plots above. Con: No color differentiation, and if there are many 0's in the data, like there are in this one, the plot isn't as interesting because the quartile ranges are collapsed with the median at and around 0.
```

```python
sns.catplot(x="staff_pick", y="usd_pledged", kind="point", data=df)
#This is a point plot. Pro: You can better see the difference in the medians and spreads between the two categories. Con: Not as common as the box plot, so not as easy for others to interpret.
```

```python
sns.catplot(x="staff_pick", y="usd_pledged", kind="violin", data=df);
#This is a violin plot. Pro: Visually more impactful than a regular box plot. Con: May be difficult to interpret if you're not as familiar with this type of plot.
```

```python
sns.catplot(x="staff_pick", y="usd_pledged", kind="bar", data=df);
#This is a bar chart. Pro: It also presents the ranges of the variables with the lines so you get a better idea of the spread of the data. Con: May not be obvious what the lines represent.
```
