<b>Should We be Concerned about "Populations of Concern" in the U.S.?</b>

If you follow national political discourse, refugees have been a recent topic of interest and source of debate. Some will point out that the U.S. has a history of welcoming refugees, assylees, and other "populations of concern." What has been the history of the U.S. in regard to the number, type, and country of origin of "populations of concern" over time? Are we seeing increasing numbers, and are they typically from the same countries or different ones? This exploratory analysis will help us understand the nature of the scope of "populations of concern" in the U.S. and suggest recommendations for future research.

The Humanitarian Data Exchange (https://data.humdata.org/) keeps data to help us answer this question. It is an extensive database interested persons may access for free to download and analyze data. The UNHCR - The UN Refugee Agency - has uploaded a dataset called "UNHCR's populations of concern residing in the United States" (https://data.humdata.org/dataset/refugees-residing-usa). It contains data on the number of Refugees, Returned Refugees, Asylum-Seekers, Internally Displaced Persons (IDPs), Returned IDPs, Stateless Persons, and Other Persons of Concerns, including the Total Population of Concern in the U.S. from 1951 to 2017.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('/Users/roxyallen/Downloads/populations_of_concern.csv')
```

Let's take a look at the big picture of Populations of Concern entering the U.S. from 1951-2017 coming from various countries. Has the number increased over time? Have there been any spikes in any of the years?

```python
plt.hist(df['Year'], bins=60)

plt.xlabel('Year')
plt.ylabel('Number of Unique Countries of Origin')
plt.title('Histogram of Number of Unique Countries of Origin Per Year')
```

Looking at the histogram above, the unique number of countries from where the Populations of Concern originate has been flat from 1951 to about 1988. The UNHCR did not keep records on which country the persons originated until the late 1980s. We then see that the number has increased, especially in the mid-1990s, with spikes in at that time period, the late 2000's, and the most recent year in 2017.

```python
plt.subplots(figsize=(10,8))
plt.boxplot(df.Year)

plt.xlabel('Year')
plt.title('Boxplot for Year')
plt.show()
```

The boxplot above demonstrates that the number of countries from where the Populations of Concern originate are, on average coming in the mid-to-late 2000's. There are many outliers representing the earlier years before the agency collected data on the specific countries from where the populations of concern originate. If we were to analyze this further, it would be worth it to remove those outliers to work with data that are more like each other.

That gives us a sense of trends over time for all of the Total Populations of Interest. But there are different types of Populations of Interest. Two of the main types are Refugees and Assylum Seekers. Is there a difference in the number of Refugees vs. Assylum seekers coming to the U.S.? Two side by side boxplots, or subplots, can help us better understand how these two groups break down.

```python
df['Refugees (incl. refugee-like situations)'] = df['Refugees (incl. refugee-like situations)'].fillna(0.0).astype(int)
df['Asylum-seekers (pending cases)'] = df['Asylum-seekers (pending cases)'].fillna(0.0).astype(int)

plt.figure(figsize=(20, 15))
x = df['Refugees (incl. refugee-like situations)']
plt.subplot(1, 2, 1)
plt.boxplot(x)
plt.ylabel('Number of Refugees')
plt.title('Boxplot of Refugees')

y = df['Asylum-seekers (pending cases)']
plt.subplot(1, 2, 2)
plt.boxplot(y)
plt.ylabel('Number of Assylum Seekers')
plt.title('Boxplot of Assylum Seekers')
plt.show()
```

The above subplots demonstrate that the data are very dispersed, with many outliders. This relates to the histogram above where we saw that in the earlier years from the 1950s to the 1980s, the UNHCR was not keeping detailed data on the country of origin, so some years the data are reported in aggregate for both the refugees and assylum seekers. However, we can see that the maximum number of refugees (over 700,000) is larger than the maximum number of assulum seekers (a bit over 17,500). We receive more refugees than assylum seekers in the U.S.

Since refugees are the largest group of Populations of Concern, what percent, on average, of refugees return home?

```python
np.mean(df['Returned refugees']) / np.mean(df['Refugees (incl. refugee-like situations)'])
```

Only 3% of refugees, on average, have returned home.

The above analyses have explored various parts of this dataset, and we can see that several of the variables have outliers due to how the data were collected by UNHCR in the earlier years of the 1950s to mid 1980s. 

We also see that there are certain years where the U.S. has welcomed people from these Populations of Concern from a larger number of different countries at certain time periods, which would be worth further exploration of which countries those were. These spikes in the data could be due to certain geo-political developments or federal policies. A bar chart would be helpful for this analysis. A bar chart would also be a great visualization of the breakdown of the different types of Populations of Concern. It would also be informative if we could predict the return of refugees from this data. For example, perhaps people from certain countries or certain types of the Populations of Concern are more likely to return home.

Overall, we do see an increase of Populations of Concern over time, with more refugees than assylum seekers, and very few refugees, on average, return home. This may be concerning to those who do not support welcoming many from the Populations of Concern; and at the same time it could be concerning to those in support that we do not support more from these populations. Further research, as described in the paragraph above, would dig into more details to help policy makers and community leaders better understand these populations and make decisions that are based on their values.
