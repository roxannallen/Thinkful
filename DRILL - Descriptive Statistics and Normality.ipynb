```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gumbel = np.random.gumbel(5,1, 1000)

#Plot a histogram.
plt.hist(gumbel)

# Print the histogram
plt.show()
```

```python
beta = np.random.beta(5,1, 1000)

#Plot a histogram.
plt.hist(beta)

# Print the histogram
plt.show()
```

```python
binomial = np.random.binomial(5,1, 1000)

#Plot a histogram.
plt.hist(binomial)

# Print the histogram
plt.show()
```

```python
chisquare = np.random.chisquare(5, 1000)

#Plot a histogram.
plt.hist(chisquare)

# Print the histogram
plt.show()
```

```python
dirichlet = np.random.dirichlet((10, 5, 3), 20)

#Plot a histogram.
plt.hist(dirichlet)

# Print the histogram
plt.show()
```

```python
exponential = np.random.exponential(5, 1000)

#Plot a histogram.
plt.hist(exponential)

# Print the histogram
plt.show()
```

```python
var1 = np.random.normal(5,.5, 100)
var2 = np.random.normal(10,1, 100)
var3 = var1+var2

mean = np.mean(var3)
sd = np.std(var3)

plt.hist(var3)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
print('Results will vary depending on random seed.  In general, stats should show some skew.')
```
